```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

hdul = fits.open('sw00010220001xpcw3po_cl.evt.gz')
x = hdul[1].data['X']
y = hdul[1].data['Y']
pi = hdul[1].data['PI']
filteredindices = np.where(pi>50)
fig, ax = plt.subplots(figsize=(16, 12))
img = plt.scatter(x[filteredindices], y[filteredindices], c=pi[filteredindices])
plt.colorbar(img)
plt.show()
print(hdul[0].header['DATE'])
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
ax.set_facecolor('black')
ax.add_artist(plt.Circle((920, 180), 395, fill=False, color='red'))
plt.scatter(x[filteredindices], y[filteredindices], color='white', s=0.4)
```

```python
plt.hist(pi[filteredindices], bins=40)
```
