```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
hdul = fits.open('events/10_00727541016_sw00727541016xpcw3po_cl.evt.gz')
xs = hdul[1].data['X']
ys = hdul[1].data['Y']
pis = hdul[1].data['PI']
plt.figure(figsize=(12, 12))
plt.scatter(xs, ys, c=pis)
#2017-01-03 10:55:08
```

```python
plt.hist(pis, bins=40)
plt.show()
```

```python
filteredindices = np.where(pis > 100)
plt.figure(figsize=(12, 12))
plt.scatter(xs[filteredindices], ys[filteredindices], c=pis[filteredindices])
```

```python
plt.hist(pis[filteredindices], bins=40)
plt.show()
```
