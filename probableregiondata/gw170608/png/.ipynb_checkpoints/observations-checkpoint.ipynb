```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
hdul = fits.open('../events/0_07003917001_sw07003917001xpcw3po_cl.evt.gz')
x = hdul[1].data['X']
y = hdul[1].data['Y']
pi = hdul[1].data['PI']
filteredindices = np.where(pi>100)
plt.hist(pi[filteredindices], bins=40)
plt.show()
plt.clf()
fig, ax = plt.subplots(figsize=(16, 12))
img = plt.scatter(x[filteredindices], y[filteredindices], c=pi[filteredindices])
plt.colorbar(img)
plt.show()
```
