```python
import pandas as pd
df = pd.DataFrame()
file = open('data.txt', 'r')
tids = []
oids = []
ras = []
decs = []
stimes = []
modes = []
for line in file.readlines():
    arr = line.split()
    tids.append(arr[0])
    oids.append(arr[1])
    ras.append(arr[2] + ' ' + arr[3] + ' ' + arr[4])
    decs.append(arr[5] + ' ' + arr[6] + ' ' + arr[7])
    stimes.append(arr[8] + ' ' + arr[9])
    modes.append(arr[12])
df['targetid'] = tids
df['obsid'] = oids
df['ra'] = ras
df['dec'] = decs
df['start time'] = stimes
df['mode'] = modes
df = df[df['mode'] == 'PHOTON'].sort_values(by=['start time'])
df
```

```python
df.to_csv('data.csv', index=False)
```

```python
import os
from shutil import copyfile
c = 1
for i, row in df.tail(74-26).head(25).iterrows():
    obsid = row['obsid']
    for file in os.listdir('events'):
        if obsid in file:
            copyfile('events/' + file, 'mosaics/fits'+str(c)+'.evt.gz')
    for file in os.listdir('expos'):
        if obsid in file:
            copyfile('expos/' + file, 'mosaics/expo'+str(c)+'.evt.gz')
    c += 1
```
