```python
import pandas as pd
df = pd.DataFrame()
file = open('data.txt', 'r')
tids = []
oids = []
ras = []
decs = []
stimes = []
for line in file.readlines():
    arr = line.split()
    tids.append(arr[0])
    oids.append(arr[1])
    ras.append(arr[2] + ' ' + arr[3] + ' ' + arr[4])
    decs.append(arr[5] + ' ' + arr[6] + ' ' + arr[7])
    stimes.append(arr[8] + ' ' + arr[9])
df['targetid'] = tids
df['obsid'] = oids
df['ra'] = ras
df['dec'] = decs
df['start time'] = stimes
df = df.sort_values(by=['ra', 'dec'])
df
```

```python
df.to_csv('../gw190814/data.csv')
```
