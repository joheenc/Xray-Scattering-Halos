```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
hdul = fits.open('mosaics/last.fits')[0].data
```

```python
xs = []
ys = []
for i in range(hdul.shape[0]):
    for j in range(hdul.shape[1]):
        if hdul[i][j] != 0:
            xs.append(i)
            ys.append(j)
ig, ax = plt.subplots(figsize=(20, 20))
ax.set_facecolor('black')
plt.scatter(xs, ys, color='white', s=5)
plt.savefig('last_mosaic.png')
```
